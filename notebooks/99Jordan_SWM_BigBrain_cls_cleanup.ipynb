```python
import numpy as np
import nibabel as nib
from skimage.segmentation import flood, flood_fill
import copy
```

```python
dir = "/host/percy/local_raid/youngeun/sub-BigBrain/Histological-Space/200um/"
```

```python
## Clean up GM and WM interface

data = nib.load(dir + "gm-wm.nii")
gm_data = nib.load(dir + "Grey_mask.nii")
gmwm = data.get_fdata()
gm = gm_data.get_fdata()

gmwm[(gmwm > 2) & (gmwm <= 3)] = 2

nib.save(nib.Nifti1Image(gmwm, gm_data.affine), dir + 'gm-wm-cleanup.nii.gz')
```

```python
data_cls = nib.load(dir + "gm-wm-sct-cleanup.nii")
gmwmsct = data_cls.get_fdata()

uni = np.unique(gmwmsct)
print(uni)
```

```python
orig = nib.load("full_cls_400um_2009b_sym.nii.gz")
gm = nib.load("Grey_mask.nii")

t = orig.get_fdata()
gmdat = gm.get_fdata()


newgm = copy.deepcopy(gmdat)
newgm[newgm==2] = 1
newgm = flood_fill(newgm,(290,310,210), new_value=2)
newgm = flood_fill(newgm,(210,310,210), new_value=2)
newgm[t==6] = 3
newgm[t==1] = 3
newgm[gmdat==1] = 1

nib.save(nib.Nifti1Image(newgm, gm.affine),'gm-wm-sct.nii.gz')
```
